Формула Эйлера для комплексных чисел.

```python
import math
import cmath

def taylor_euler(x, n_terms=17):
    sum_euler = 0j  # Инициализируем комплексным нулем (0 + 0j)
    for n in range(n_terms):
        # Формула члена ряда Тейлора для e^(ix): (i*x)^n / n!
        term = ((1j * x) ** n) / math.factorial(n)
        sum_euler += term
    return sum_euler

x_val = float(input("Введите значение x = "))
print(f"Приближенное значение e^(i*{x_val}): {taylor_euler(x_val)}")
print(f"Точное значение e^(i*{x_val}): {cmath.exp(1j * x_val)}")
```
